## Worksheet 20 Question 2

Find the circulation and flux of 
$$
\boldsymbol{F} = x \boldsymbol{i} + y \boldsymbol{j}
$$
around/across the curve
$$
\boldsymbol{r}(t) = (3 \cos t) \boldsymbol{i} + (4 \sin t) \boldsymbol{j}, \, t \in [0, 2 \pi]
$$

**Solution**

The circulation is 
\begin{align*}
& \int_C \boldsymbol{F} \cdot d \boldsymbol{r} \\
= & \int_{0}^{2 \pi} \boldsymbol{F} \cdot  \boldsymbol{r}' d t \\
= & \int_{0}^{2 \pi} ( 3 \cos t, 4 \sin t) \cdot (-3 \sin t, 4 \cos t) dt \\
= & \int_{0}^{2 \pi} 7 \sin t \cos t dt \\
= & \frac{7}{2} \int_{0}^{2 \pi} \sin(2 t) dt = 0
\end{align*}

The flux is 
\begin{align*}
& \int_C M d y - N dx \\
= & \int_{0}^{2 \pi} (M y' - N x') d t\\
= & \int_{0}^{2 \pi} ( (3 \cos t) (4 \cos t) - (4 \sin t) ( - 3 \sin t) )d t \\
= & \int_{0}^{2 \pi} 12 (\cos^2 t + \sin^2 t) dt \\
= & \int_{0}^{2 \pi} 12 dt = 24 \pi
\end{align*}

## Line integral versus integral of a vector field along a curve

A good reference for the difference between line integral versus integral of a vector field along a curve is the following article by Terance Tao

http://www.math.ucla.edu/~tao/preprints/forms.pdf

The line integral correspondes to the Lebesgue integral (unsigned definite integral for a single variable) and the integral of a vector field is more closely related to the integral of a differential form (signed definite integral for a single variable). 

The **line integral** (or more generally the Lebesgue integral) is closely related to the idea of generalizing sums. Suppose we divide our curve $C$ into many small segments, and we want to sum up the produce of function values with segment length
$$
\sum_{i=1}^{N} f(x_i, y_i) \Delta s_i
$$

We then let $N$ goes to infinity and make the segments infinitesmally short (intuitively), we get the line integral
$$
\int_C f(x, y) ds
$$

The lower and upper limits don't even show up in this process, so naturally it doesn't depend on the direction of the curve. 

However, when we want to compute it, we need switch to a parameter form of the curve and lower and upper limit show up. In the notation of the above reference, this gives us the following unsigned definite integral
$$
\int_{[a, b]} f(x(t), y(t)) |r'(t)| dt 
$$
Here we just take $[a, b]$ as a set. To ensure we got positive meausre, we will place the smaller number on the lower limit, and larger number on the upper limit. 

(If we are integrating functions over a surface, or more generally, a manifold, we can still write it as 
$$
\int_S f(x, y, z) dA
$$
where $S$ is the surface (as a set) and $dA$ is the infinitesmal area element, and orientation of the surface still doesn't matter. When we switch to parameter form of the surface, we use the correct area form to correctly handle the lower and upper limits). 

For **the integral of a vector field along a curve**, it is more closely related to the integral of a differential form, and the integral of a differential form actually depends on the orientation of the curve (or more generally, the orientation of the manifold). (If we study the integration of a differential form, we see it is essential for the manifold to have an orientation for us to define such a thing). 





## Worksheet 18, Question 3

Let $R$ be the region in the first quadrant that is bounded by hyperbolas $xy = 4$, $xy = 16$ and the lines $y = x$, $y = 4x$. Use the transformation 
$$
x = u/v, \, y = uv
$$
with $u > 0, v > 0$ to rewrite 
$$
\int \int_R \left( (y/x)^2 + (xy)^{-1} \right) d x dy
$$
as an integral over a region $G$ in the $uv$-plane. Hence evaluate the integral. 

**Solution**

We first figure out the transformed region $G$. 

Solve for $u, v$ in terms of $x, y$. Note $u^2 = xy \Rightarrow u = \sqrt{xy}$ (note we use $u>0$), and $v^2 = y/x \Rightarrow v = \sqrt{y/x}$ (note we use $v > 0$). 

For the boundaries of $R$, we have: 
$$
xy = 4 \Leftrightarrow u = 2, \, xy = 16 \Leftrightarrow u = 4
$$
$$
y = x \Leftrightarrow v = 1, \, y = 4x \Leftrightarrow v = 2
$$
So the region $G$ is 
$$
G: 2 \leq u \leq 4, \, 1 \leq v \leq 2
$$

Second, we compute the Jacobian: 
$$
\frac{\partial x}{\partial u} = 1/v, \, \frac{\partial x}{\partial v} = -u/v^2, \, \frac{\partial y}{\partial u} = v, \, \frac{\partial y}{\partial v} = u, 
$$
So 
$$
\frac{\partial (x, y)}{\partial (u, v)} = \frac{2 u}{v}
$$

Third, we write integrand in terms of $u, v$
$$
f(x, y) = (y/x)^2 + (xy)^{-1} = v^4 + 1/u^2
$$

Compute the integral
$$
\int \int_R \left( (y/x)^2 + (xy)^{-1} \right) d x dy = \int \int_G \left( v^4 + \frac{1}{u^2} \right) \frac{2 u}{v} d u dv = 45 + 2 (\ln 2)^2
$$

### Question 3
Find the centroid of the region $D$ which consists of that part of the paraboloid $z = x^2 + y^2 - 4$ below the $xy$-plane. 

**Solution**

Please sketch the region by yourself.

We discussed in recitation that we can solve it in a similar way to question 2, the following are the details of this method. See the "**alternatively**" part for a more convenient solution that we didn't have time to discuss in recitation. 

Here we note that the paraboloid intersects with $xy$-plane when $z = 0$, which gives $x^2 + y^2 = 4$. Lower bound of $z$ is given by the paraboloid equation, and upper bound for $z$ is given by $z = 0$. And the region can be described as: 
$$
D: x^2 + y^2 \leq 4, x^2 + y^2 - 4 \leq z \leq 0
$$
which can also be written as:
$$
D: -2 \leq x \leq 2, - \sqrt{4 - x^2} \leq y \leq \sqrt{4 - x^2}, x^2 + y^2 - 4 \leq z \leq 0
$$

Using these bounds, we can evaluate the integrals in the order of $dz dy dx$ and find $\bar{x}, \bar{y}$ and $\bar{z}$. 

Note the region $D$ is symmetric in $x$ and $y$, so the $x$ and $y$ coordinates of the centroid are just $0$: 
$$
\bar{x} = 0, \bar{y} = 0
$$
One can confirm this by evaluating the corresponding integrals. 

Now 
$$
\begin{align*}
& \int_{-2}^{2} \int_{-\sqrt{4 - x^2}}^{\sqrt{4 - x^2}} \int_{x^2 + y^2 - 4}^{0} z \, dz dy dz \\
& = \int_{-2}^{2} \int_{-\sqrt{4 - x^2}}^{\sqrt{4 - x^2}} - \frac{1}{2} \left( x^2 + y^2 - 4 \right)^2 \, dy dx 
\end{align*}
$$
To evaluate this double integral, it is best to switch the polar coordinates, since now our region is just the interior of the circle $x^2 + y^2 = 4$ and the integrand is $- \frac{1}{2} \left( x^2 + y^2 - 4 \right)^2  = - \frac{1}{2} \left( r^2 - 4 \right)^2$. So switch to polar coordinates, we get: 
$$
\begin{align*}
& \int_{-2}^{2} \int_{-\sqrt{4 - x^2}}^{\sqrt{4 - x^2}} - \frac{1}{2} \left( x^2 + y^2 - 4 \right)^2 \, dy dx \\
& = \int_{0}^{2 \pi} \int_{0}^{2} - \frac{1}{2} \left( r^2 - 4 \right)^2 r \, dr d\theta \\
& = -\frac{32 \pi}{3}
\end{align*}
$$
Similarly, we can find the volume 
$$
\int_{-2}^{2} \int_{-\sqrt{4 - x^2}}^{\sqrt{4 - x^2}} \int_{x^2 + y^2 - 4}^{0} 1 \, dz dy dz = 8 \pi
$$

So 
$$
\bar{z} = \frac{-\frac{32 \pi}{3}}{8 \pi} = -\frac{4}{3}
$$

**Alternatively**, we can evaluate the integrals in the order of $dx dy dz$. This is a little bit unusual, but actually is very convenient in this problem, since each $z$ cross-section is simply a circular disk. And we have: 
$$
\begin{align*}
& \int \int \int_D z d V \\
& = \int_{-4}^{0} \int_{x^2 + y^2 \leq 4 + z} z \, dx dy dz\\
& = \int_{-4}^{0} z \int_{x^2 + y^2 \leq 4 + z}  \, dx dy dz
\end{align*}
$$

Note that $\int_{x^2 + y^2 \leq 4 + z}  \, dx dy$ is the area of a circular disk with radius $\sqrt{4 + z}$, so it is just $\pi ( 4 + z)$. Then we continue: 
$$
\int \int \int_D z d V = \int_{-4}^{0} z  \pi ( 4 + z) dz = -\frac{32 \pi}{3}
$$
Similarly, we have 
$$
\int \int \int_D  d V = \int_{-4}^{0} \pi ( 4 + z) dz = 8 \pi 
$$

$$
I = \int_{-1}^{1} \int_{x^2}^{1} \int_{0}^{1 - y} dz dy dx 
$$

(b) Rewrite the integral in the order of $dx dy dz$ and evaluate it. 

**Solution**

We talked about finding out the bounds using figures in the recitation. Here I will discuss the algebraic way to find those bounds. A remark is that the algebraic way is not always easy, and if we have a figure, I will recommend using the figure to find the bounds since it is usually easier. 

Our domain is $D = \{(x, y, z): -1 \leq x \leq 1, x^2 \leq y \leq 1, 0 \leq z \leq 1 - y \}$. 

(b) In the order of $dx dy dz$, we first figure out the outer most variable's bounds, that is the lower and upper bounds for $z$. This can be formulated into a constraint minimization and maximization problem. Let $f(x, y, z) = z$, then the lower bound for $z$ is given the solution to the following optimization problem: 
$$
\min f(x, y, z)
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1, 0 \leq z \leq 1 - y
$$
In the current case, this is trivial, the min value is $0$, given by the third inequality. 

The upper bound for $z$ is given by the solution to: 
$$
\max f(x, y, z)
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1, 0 \leq z \leq 1 - y
$$

In general, solving a constrained optimization is a highly nontrivial task. But in the current case, we can proceed in the following steps: 

First, $f(x, y, z) = z \leq 1 - y$, so its max value is always bounded by $1-y$. In order to make this bound as high as possible, we need to make $y$ as small as possible. Note $y \geq x^2$, so the lowest value of $y$ is $x^2$. Plug this into the upper bound $1 - y$, we get $1 - x^2$. Now in order to make this upper bound as high as possible, we need to make $x^2$ as small as possible. Note $-1 \leq x \leq 1$, so the lowest value of $x^2$ is $0$. Plug this into $1 - x^2$, we get $1$. Now we see that a upper bound for $f(x, y, z)$ is $1$. It is not yet a max value, since the upper bound may be not obtainable. So we check if we can assign proper values to ensure this is obtained in $D$. Indeed, we can take $x = 0, y = 0, and z = 1$ to achieve this upper bound. So the max value of $f$ is $1$. 

Therefore, the bounds for $z$ integral is $0$ and $1$. 

Now, we figure out bounds for $y$. Fix a $z \in [0, 1]$, the lower bound for $y$ is given by 
$$
\min y
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1, 0 \leq z \leq 1 - y
$$

Note this time $z$ is a fixed value between $0$ and $1$, so we can't change its value. 

First rewrite the third inequality into $y \leq 1 - z$, and note then the second and this new inequality can be combined into $x^2 \leq y \leq \min(1, 1 - z)$. Also note since $z \in [0, 1]$, we have $1 - z \leq 1$, so $\min(1, 1-z) = 1-z$. So the optimization problem becomes: 
$$
\min y
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1 - z
$$

Now a lower bound for $y$ is $x^2$, and to make this as small as possible, we take $x = 0$, and get the min value of $y$ being $0$. 

Similarly, the upper bound is given by: 
$$
\max y
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1 - z
$$
Clearly the upper bound is $1 - z$. 

Therefore, the $y$ integral has bounds $0$ and $1-z$. 

Finally, for $x$. Fix a $y$ and a $z$, we solve
$$
\min x
$$
such that 
$$
-1 \leq x \leq 1, x^2 \leq y \leq 1, 0 \leq z \leq 1 - y
$$
Note now that $y$ and $z$ are fixed and the third inequality has nothing to do with $x$, it poses no constraint on $x$ and we simply ignore it. From $x^2 \leq y$, we get $-\sqrt{y} \leq x \leq \sqrt{y}$. Combine this with the first inequality, we get: 
$$
\max(-1, -\sqrt{y}) \leq x \leq \min(1, \sqrt{y})
$$
Recall $0 \leq y \leq 1 - z$ and $0 \leq z \leq 1$, so $\sqrt{y} \leq 1$, so $\sqrt{y} \leq 1$ and $-\sqrt{y} \geq -1$. So we get: 
$$
-\sqrt{y} \leq x \leq \sqrt{y}
$$
which specifies the bounds for $x$ integral. 

Therefore we get: 
$$
I = \int_{0}^{1} \int_{0}^{1-z} \int_{-\sqrt{y}}^{\sqrt{y}} dx dy dz
$$
Evaluating it, we have: 
$$
I = \int_{0}^{1} \int_{0}^{1-z} 2 \sqrt{y} dy dz = \int_{0}^{1} \frac{4}{3} (1 - z)^{3/2} dz = \left[- \frac{4}{3}\frac{2}{5} (1 - z)^{5/2}\right]_{0}^{1} = \frac{8}{15}
$$




## Worksheet 15 Problem 3

Find the area of the region common to the interiors of the cadioids $r = 1 + \cos \theta$ and $r = 1 - \cos \theta$. 

**Solution**

Denote the region by $R$, and denote the part of the region that is in the first quadrant by $R_1$. As we discussed in recitation, by symmetry, we know $\text{Area}(R) = 4 \, \text{Area}(R_1)$. Therefore, we first need to evaluate $\text{Area}(R_1)$. 
$$
\text{Area}(R_1) = \int \int_{R_1} d x d y = \int \int_{R_1} r d r d \theta
$$
To figure out the bounds for $r$ and $\theta$, we notice that $R_1$ is bounded by the $y$ axis and the part of $r = 1 - \cos \theta$ that is in the first quadrant. Especially, we see that for different $\theta$, the range of $r$ is different. So we take $\theta$ as the outer integer and $r$ as the inner integral. Now the range for $\theta$ in $R_1$ is from $0$ to $\pi/2$, and for a fixed $\theta$, the range for $r$ is from $0$ to $1 - \cos \theta$. So 
$$
\text{Area}(R_1) = \int_{0}^{\pi/2} \int_{0}^{1 - \cos \theta} r d r d \theta = \int_{0}^{\pi/2} \frac{ (1 - \cos \theta)^2}{2} d \theta = \frac{1}{2} \int_{0}^{\pi/2} \left( 1 - 2 \cos \theta + \cos^2 \theta \right) d \theta
$$
To integrate $\cos^2 \theta$, we use double angle formula $\cos 2 \theta = 2 \cos^2 \theta - 1 \Rightarrow \cos^2 \theta = ( 1 + \cos 2 \theta ) /2$. So 
$$
\text{Area}(R_1) = \frac{1}{2} \int_{0}^{\pi/2} \left( 1 - 2 \cos \theta + \frac{ 1 + \cos 2 \theta}{2} \right) d \theta = \frac{1}{2} \left( \frac{\pi}{2} - 2 + \frac{\pi}{4} + \frac{1}{4} \cdot 0 \right) = \frac{3 \pi}{8} - 1
$$
And 
$$
\text{Area}(R) = 4 \, \text{Area}(R_1) = \frac{3 \pi}{2} - 4
$$

## Question 2. 
Find the points on $z = xy + 1$ that are closest to the origin. 

Solution: 

Instead of maximizing the distance, we maximize the square of distance to the origin. The problem is written as: 
$$
\text{ maximize } f(x, y, z) = x^2 + y^2 + z^2
$$
subject to 
$$
g(x, y, z) = xy - z + 1 = 0
$$

Introduce Lagrange multipler $\lambda$ for $g$, we need to solve: 
$$
\nabla f = \lambda g
$$
and 
$$ 
g = 0
$$ 
which are: 
$$
2x = \lambda x, 2 y = \lambda y, 2 z = -\lambda, xy - z + 1 = 0
$$
From the first equation, we get: $x = \frac{\lambda}{2} y$, plug into the second equation, we get
$$
2 y = \frac{\lambda^2}{2} y = 0 \Rightarrow ( 2 - \frac{\lambda^2}{2} ) y = 0 \Rightarrow y = 0 \text{ or } \lambda = 2 \text{ or } \lambda = -2
$$
We discuss all these 3 cases: 

1. If $y = 0$, then $x = \frac{\lambda}{2} y = 0$, plug into the last equation $xy - z + 1 = 0$, we get $z = 1$. So we get a point $(0, 0, 1)$. 

2. If $\lambda = 2$, then $z = - \frac{\lambda}{2} = -1$, $x = \frac{\lambda}{2} y = y$, plug these two results into the last equation $xy - z + 1 = 0$, we get $x^2 + 2 = 0$ which has no solution. So $\lambda = 2$ gives us no solution. 

3. If $\lambda = -2$, then $z = - \frac{\lambda}{2} = 1$, $x = \frac{\lambda}{2} y = -y$, plug these two results into the last equation $xy - z + 1 = 0$, we get $x^2 = 0$, we get $x = 0$ and $y = 0$, so again we get the point $(0, 0, 1)$. 

So we conclude the only solution is $(0, 0, 1)$. And this is the point on the surface that is the closest to the origin, its distance is $\sqrt{1} = 1$.  

Remark: If we want to make sure that $(0, 0, 1)$ is the closest point (instead of the furthest point), we can check the distance of a different point on the surface, for example, take $x = y = 1, z = 2$ (which is on the surface), we see the distance is $\sqrt{1 + 1 + 4} = \sqrt{6} > 1$. So $(0, 0, 1)$ cannot be the furthest point. 


## Question 3
Find the max and min values of 
$$
f(x, y, z) = x - 2y + 5z 
$$
on the sphere
$$
x^2 + y^2 + z^2 = 30
$$

Solution:

We rewrite the constraint into 
$$
g(x, y, z) = x^2 + y^2 + z^2 - 30 = 0
$$

Introducing Lagrange multiplier $\lambda$ for $g$, we need to solve
$$
\nabla f = \lambda \nabla g
$$
and 
$$
g = 0
$$
which are
$$
1 = \lambda \cdot 2x, -2 = \lambda \cdot 2y, 5 = \lambda \cdot 2 z, x^2 + y^2 + z^2 - 30 = 0
$$
From the first three equations, we get: 
$$
x = \frac{1}{2 \lambda }, y = - \frac{1}{\lambda}, z = \frac{5}{2 \lambda}
$$
Plug these into the last equation, we get: 
$$
\left( \frac{1}{4} + 1 + \frac{25}{4} \right) \frac{1}{\lambda^2} = 30 \Rightarrow \lambda^2 = \frac{1}{4} \Rightarrow \lambda = \pm \frac{1}{2}
$$

If $\lambda = \frac{1}{2}$, we get $x = 1, y = -2, z = 5$, and $f(1, -2, 5) = 30$. 

If $\lambda = - \frac{1}{2}$, we get $x = -1, y = 2, z = -5$, and $f(-1, 2, -5) = -30$. 

So the max value of $f$ is $30$ and it occurs at $(1, -2, 5)$, and the min value of $f$ is $-30$ and it occurs at $(-1, 2, -5)$.
