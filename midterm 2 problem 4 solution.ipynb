## Math 2552 (Spring 2016) Midterm 2 Problem 4

The problem is: 

Use the variation of parameters to find general solutions of 
$$
y'' - 2 y' + y = \frac{e^t}{t^2}, \, (t > 0)
$$
You may use the factor that $\lambda^2 - 2 \lambda + 1 = (\lambda - 1)^2$. 

### Method-1: 

We use the variation of parameters formula on the text book. (I re-write it in a slightly different way, but it is really the same formula, and you can use the one on the textbook if you would like). 

Step-1. Find the complementary solution:
Solve $y'' - 2 y' + y = 0$, and get
$$
y_c(t) = c_1 e^t + c_2 t e^t
$$
Note we get two linearly independent solutions $y_1 = e^t$ and $y_2 = t e^t$. 

Step-2. Find a particular solution for the nonhomogeneous equation. 
First, compute the Wronskian of $y_1, y_2$: 
$$
\begin{align*}
W(t) = \mathrm{det} \begin{pmatrix}
y_1 & y_2 \\
y_1' & y_2' 
\end{pmatrix} = \mathrm{det} \begin{pmatrix}
e^t & t e^t \\
e^t & e^t + t e^t
\end{pmatrix} = e^{2 t}
\end{align*}
$$
Then compute $W_1(t)$ and $W_2(t)$: 
$$
\begin{align*}
W_1(t) = \mathrm{det} \begin{pmatrix}
0 & y_2 \\
\frac{e^t}{t^2} & y_2' 
\end{pmatrix} = \mathrm{det} \begin{pmatrix}
0 & t e^t \\
\frac{e^t}{t^2} & e^t + t e^t
\end{pmatrix} = -\frac{e^{2 t}}{t}
\end{align*}
$$
$$
\begin{align*}
W_2(t) = \mathrm{det} \begin{pmatrix}
y_1 & 0 \\
y_1' & \frac{e^t}{t^2}
\end{pmatrix} = \mathrm{det} \begin{pmatrix}
e^t & 0 \\
e^t & \frac{e^t}{t^2}
\end{pmatrix} = \frac{e^{2 t}}{t^2}
\end{align*}
$$
Next, we have: 
$$
u_1' = \frac{W_1}{W} = -\frac{1}{t}
$$
$$
u_2' = \frac{W_1}{W} = \frac{1}{t^2}
$$
and so
$$
u_1 = - \int \frac{1}{t} dt = - \ln(t)
$$
$$
u_2 = \int \frac{1}{t^2} dt = - \frac{1}{t}
$$
And the particular solution is: 
$$
y_p(t) = u_1 y_1(t) + u_2 y_2(t) = - e^t \ln(t) + (-1/t) t e^t = -e^t \ln t - e^t
$$

Step-3: We write out the general solution: 
$$
y(t) = y_c(t) + y_p(t) = c_1 e^t + c_2 t e^t -e^t \ln t - e^t
$$



### Method-2: 
We first re-write it as a system of two linear equations, and then solve using the variation of parameters for a system of two equations. I will be concise on this, since it is the same method used by a few textbook examples: 

step-1: we solve for $y_c(t)$ and get $y_c(t) = c_1 e^t + c_2 t e^t$; 

step-2: We re-write the equation into a system of two equations by letting $x_1 = y$, $x_2 = y'$, and get: 
$$
\begin{align*}
& x_1' = x_2\\
& x_2' = -x_1 + 2 x_2 + \frac{e^t}{t^2}
\end{align*}
$$
The fundamental matrix of this system is: 
$$
\begin{eqnarray*}
M(t) = \begin{pmatrix}
e^t & t e^t \\
e^t & e^t + t e^t 
\end{pmatrix}
\end{eqnarray*}
$$
We have 
$$
\frac{d \vec{u}}{d t} = M^{-1}(t) \begin{pmatrix} 0 \\ \frac{e^t}{t^2}  \end{pmatrix} = \begin{pmatrix} - 1/t \\ 1/t^2 \end{pmatrix}
$$
So 
$$
\vec{u}(t) = \int \frac{d \vec{u}}{d t}  dt = \begin{pmatrix} -\ln t  \\ -1/t \end{pmatrix}
$$
And a particular solution for the system is
$$
\vec{x}_p(t) = M(t) u(t) = \begin{pmatrix} 
-e^t \ln t - e^t\\
-e^t \ln t - e^t - e^t/t
\end{pmatrix}
$$
Taking the first component, we get: 
$$
y_p(t) = -e^t \ln t - e^t
$$

Step-3: We write out the general solution: 
$$
y(t) = y_c(t) + y_p(t) = c_1 e^t + c_2 t e^t -e^t \ln t - e^t
$$
