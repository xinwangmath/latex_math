I will try to explain how to draw a quadric surface more generally. A list of all the quadric surfaces can be found at [http://mathworld.wolfram.com/QuadraticSurface.html]. I will only mention a few common ones. 

If the equation doesn't contain all variables $x$, $y$ and $z$, then its name usually ends with a "cylinder" (except for the degenrate case of intersecting planes). For example: $x^2/a^2 + y^2/b^2 = 1$ is an elliptic cylinder, $x^2/a^2 - y^2/b^2 = 1$ is a hyperbolic cylinder. To draw these two surface, we simple need to draw an ellipse or a hyperbola in the $x-y$ plane, then extend that parallel to the $z$ axis (in these two examples, it is the $z$ axis, if the equation is different, it could be the $x$ or $y$ axis). 

If the equation contains all variables $x$, $y$ and $z$, and all of them shows up as square terms, then it is either an ellipsoid or a hyperboloid of one or two sheets. 

$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1 \text{ ellipsoid.} $$ 
$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1 \text{ hyperboloid of one sheet.} $$ 
$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1 \text{ hyperboloid of one sheet.} $$ 

If the equation contains all variables $x$, $y$ and $z$, and one of them shows up in linear term, then it is a paraboloid. 

$$ z = \frac{x^2}{a^2} + \frac{y^2}{b^2} \text{ ellipic paraboloid } $$
$$ z = \frac{x^2}{a^2} - \frac{y^2}{b^2} \text{ hyperbolic paraboloid } $$

Plug this to the $x$ components, we see they are also equal, therefore, this vector equation has a solution. So they will collid.
